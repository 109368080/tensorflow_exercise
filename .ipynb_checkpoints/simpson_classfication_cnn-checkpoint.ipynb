```python
x_train, y_train = simpson_img, label
print('input_data shape: training {training_shape}'.format(
            training_shape = x_train.shape))
print('y_true shape: training {training_shape}'.format(
            training_shape = y_train.shape))
```

```python
def weight_variable(shape):
    initial=tf.truncated_normal(shape, mean=0.0, stddev=0.1)
    return tf.Variable(initial)

def bias_variable(shape):
    initial=tf.constant(0,1,shape=shape)
    return tf.Variable(initial)

def conv2d(x, W):
    # stride [1, x_movement, y_movement, 1]
    # Must have strides[0] = strides[3] = 1
    return tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')

def max_pool_2x2(x):
    # stride [1, x_movement, y_movement, 1]
    return tf.nn.max_pool(x, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
```

```python
tf.reset_default_graph()
```

```python
input_data = tf.placeholder(tf.float32, [None, 50, 50, 3])
print(input_data)
y_true = tf.placeholder(tf.float32, [None, 6])
# x_image = tf.reshape(input_data, [-1, 50, 50, 3])
# print(x_image.shape)  # [n_samples, 28,28,1]

## conv1 layer ##
W_conv1 = weight_variable([5, 5, 3, 32]) # patch 5x5, in size 1, out size 32
b_conv1 = bias_variable([32])
h_conv1 = tf.nn.relu(conv2d(input_data, W_conv1) + b_conv1) # output size 28x28x32
h_pool1 = max_pool_2x2(h_conv1)                                         # output size 14x14x32

## conv2 layer ##
W_conv2 = weight_variable([5,5, 32, 64]) # patch 5x5, in size 32, out size 64
b_conv2 = bias_variable([64])
h_conv2 = tf.nn.relu(conv2d(h_pool1, W_conv2) + b_conv2) # output size 14x14x64
h_pool2 = max_pool_2x2(h_conv2)                                         # output size 7x7x64

## fc1 layer ##
W_fc1 = weight_variable([13*13*64, 1024])
b_fc1 = bias_variable([1024])
h_pool2_flat = tf.reshape(h_pool2, [-1, 13*13*64])
h_fc1 = tf.nn.relu(tf.matmul(h_pool2_flat, W_fc1) + b_fc1)
# h_fc1_drop = tf.nn.dropout(h_fc1, keep_prob)

## fc2 layer ##
W_fc2 = weight_variable([1024, 6])
b_fc2 = bias_variable([6])

y_conv = tf.nn.softmax(tf.matmul(h_fc1,W_fc2)+b_fc2)
# fc1 = tf.layers.dense(h_pool2, 512)
# fc2 = tf.layers.dense(fc1, 6)

y_pred = y_conv 
# prediction = tf.nn.softmax(tf.matmul(h_fc1_drop, W_fc2) + b_fc2)
```

```python
# 定義 Loss
# loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels = y_true, logits = y_pred))
loss = -tf.reduce_sum(y_true*tf.log(y_pred))
```

```python
# 定義優化器
opt = tf.train.GradientDescentOptimizer(learning_rate = 0.001)
update = opt.minimize(loss)
```

```python
y_pred_out = tf.arg_max(y_pred,1)
bool_pred = tf.equal(tf.arg_max(y_true,1),y_pred_out)
accuracy = tf.reduce_mean(tf.cast(bool_pred,tf.float32)) # 准确率
```

```python
#檢查Variables
tf.global_variables()
```

```python
from tqdm import tqdm #用來顯示進度條的套件
from sklearn.metrics import accuracy_score

def generatebatch(X, Y, n_examples, batch_size):
    for batch_i in range(n_examples // batch_size):
        start = batch_i*batch_size
        end = start + batch_size
        batch_xs = X[start:end]
        batch_ys = Y[start:end]
        yield batch_xs, batch_ys
        
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

epoch = 30 #epoch
bs = 32 #batch size
update_per_epoch = 100 #一個epoch要跑幾個batch

tr_loss = list() #存training過程中的loss值
tr_acc = list() #存training過程中的準確率

for i in range(epoch):
    training_loss = 0
    training_acc = 0
    bar = tqdm(range(update_per_epoch))
    
    for j in tqdm(range(update_per_epoch)):
        for batch_xs,batch_ys in generatebatch(x_train, y_train, y_train.shape[0], 32): # 每个周期进行MBGD算法
            tr_pred, training_loss_batch, _ = sess.run([y_pred, loss, update],feed_dict={input_data:batch_xs,y_true:batch_ys})
#         indices = np.random.choice(x_train.shape[0], 32)
#         images_batch = x_train[indices]
#         labels_batch = y_train[indices]
#         print(labels_batch)
#         tr_pred, training_loss_batch, _ = sess.run([y_pred, loss, update], feed_dict = {
#             input_data : x_train,
#             y_true : y_train
#         })
            training_loss += training_loss_batch
            training_acc_batch = accuracy_score(np.argmax(batch_ys, axis=1), np.argmax(tr_pred, axis=1))
            training_acc += training_acc_batch
        
        if j % 5 == 0:
            bar.set_description('loss: %.4g' % training_loss_batch) #進度條
            
    training_loss /= update_per_epoch
    training_acc /= update_per_epoch
    
    tr_loss.append(training_loss)
    tr_acc.append(training_acc)
    
    print('epoch {epochs}: training loss {training_loss}'.format(
            epochs = (i+1), 
            training_loss = training_loss))#每個epoch結束後顯示目前的的training loss
```
