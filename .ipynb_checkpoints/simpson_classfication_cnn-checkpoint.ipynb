```python
from sklearn.utils import shuffle

def simpson_train_batch_generator(x, y, bs, shape):
    x_train = np.array([]).reshape((0, shape, shape, 3))
    y_train = np.array([]).reshape((0, y.shape[1]))
    while True:
        new_ind = shuffle(range(len(x)))
        x = x.take(new_ind)
        y = np.take(y, new_ind, axis = 0)
        for i in range(len(x)):
#             dir_img = '../simpsons_dataset/' + x.img.iloc[i]
#             img = cv2.imread(dir_img, 0)
#             img = cv2.resize(img, (50,50))
            x_train = np.row_stack([x_train, simpson_img])
            y_train = np.row_stack([y_train, y[i]])
            if x_train.shape[0] == bs:
                x_batch = x_train.copy()
                x_batch /= 255.
                y_batch = y_train.copy()
                x_train = np.array([]).reshape((0 , shape, shape, 3))
                y_train = np.array([]).reshape((0 , y.shape[1]))        
                yield x_batch, y_batch
```

```python
# 建立placeholder
input_data = tf.placeholder(dtype = tf.float32, shape = [None, 50, 50, 3], name = 'input_data')
y_true = tf.placeholder(dtype = tf.float32, shape = [None,  y_train.shape[1]], name = 'y_true')
```

```python
# 建立CNN神經網路
conv1 = tf.layers.conv2d(
    inputs = input_data,
    filters=32,
    kernel_size=[5, 5],
    padding="same",
    activation=tf.nn.relu)
pool1 = tf.layers.max_pooling2d(inputs = conv1, pool_size=[2, 2], strides=2)
conv2 = tf.layers.conv2d(
      inputs = pool1,
      filters = 64,
      kernel_size = [5, 5],
      padding = "same",
      activation = tf.nn.relu)
pool2 = tf.layers.max_pooling2d(inputs = conv2, pool_size = [2, 2], strides = 2)
pool2_flat = tf.reshape(pool2, [-1, 7 * 7 * 64])
fc1 = tf.layers.dense(pool2_flat, 512, name = 'fc1')
fc2 = tf.layers.dense(pool2_flat, y_train.shape[1], name = 'fc2')
y_pred = fc2
```

```python
# 定義 Loss
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels = y_true, logits = y_pred))
```

```python
# 定義優化器
opt = tf.train.GradientDescentOptimizer(learning_rate=0.001)

update = opt.minimize(loss)
```

```python
#檢查Variables
tf.global_variables()
```

```python
def get_batch_data(x, y):
#     label, images = generate_data()
    input_queue = tf.train.slice_input_producer([x, y], shuffle=False)
    image_batch, label_batch = tf.train.batch(input_queue, batch_size = 32, num_threads = 1, capacity = 64)
    return image_batch, label_batch
```
